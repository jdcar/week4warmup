```python
#first, the cell data needed to be trimmed and there was inconsistent capitalization
#strip() trimmed the data, lower() made it all lower case
cb1 = cb_normalized['subject'].str.strip(";").str.lower()

print (cb1)
```

```python
#There are multiple subjects in each cell, so the cells needed to be split by delimiter using explode
#There are now 317 cells which contains dups
cb2 = cb1.str.split(';')
df_exploded = cb2.explode('subjects')
print (df_exploded)
```

```python
#drop_duplicates() removes the dups
df_exploded.drop_duplicates()
df4 = df_exploded.drop_duplicates()
#len() counts the rows
num_rows = len(df4.index)
print(num_rows)
```

```python
#answer to H4 using pd
"The facet that shows up the most is " + cb_columns[3] + ". It is used "  + str(num_rows) + " times on my selected site."
```
